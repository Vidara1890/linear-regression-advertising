# 01 - Data Loading and Initial Exploration

Dataset: **Advertising** — TV, Radio, Newspaper ad spend (in thousands of dollars) and resulting Sales (in thousands of units) for 200 markets.

Goal: understand the shape and quality of the data before modeling.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv"
df = pd.read_csv(url, index_col=0)
df.head()
```

## Basic shape and info

```python
print("Shape:", df.shape)
df.info()
```

```python
df.describe()
```

## Check for missing values

```python
df.isnull().sum()
```

**Observation:** No missing values. 200 rows, 4 numeric columns (TV, Radio, Newspaper, Sales). Data looks clean and ready for analysis.

## Quick distribution check

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 3))
for ax, col in zip(axes, df.columns):
    ax.hist(df[col], bins=20, color='steelblue', edgecolor='white')
    ax.set_title(col)
plt.tight_layout()
plt.savefig("../outputs/01_distributions.png", dpi=100)
plt.show()
```

Save the cleaned dataframe for use in the next notebook.

```python
df.to_csv("../data/advertising_clean.csv")
print("Saved cleaned data.")
```
