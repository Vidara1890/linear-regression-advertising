# 04 - Model Evaluation

Evaluate the fitted linear regression model on the held-out test set.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

y_test = np.load("../outputs/y_test.npy")
y_pred = np.load("../outputs/y_pred_test.npy")
```

## Metrics

```python
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

print(f"R2:   {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE:  {mae:.4f}")
```

**Interpretation:** R2 close to 1 means the model explains most of the variance in Sales. RMSE/MAE are in the same units as Sales (thousands of units), giving a sense of typical prediction error.

## Predicted vs Actual

```python
plt.figure(figsize=(5, 5))
plt.scatter(y_test, y_pred, alpha=0.7)
lims = [min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())]
plt.plot(lims, lims, 'r--', label='Perfect prediction')
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Predicted vs Actual (Test Set)")
plt.legend()
plt.tight_layout()
plt.savefig("../outputs/04_predicted_vs_actual.png", dpi=100)
plt.show()
```

## Summary

- TV advertising spend is the strongest driver of Sales, followed by Radio.
- Newspaper spend shows little to no relationship with Sales in this dataset.
- The simple linear model achieves a strong fit on held-out data, confirming the relationship is close to linear for this dataset.
- Next step (future work): try a model with a TV×Radio interaction term to capture the mild non-linearity seen in the residuals.
